```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Membaca data
data_tips = pd.read_csv('tips.csv')
data_tips.head()
```

```python
# Menampilkan info
data_tips.info()
data_tips.describe()
```

```python
# Visualisasi data
plt.title('Scatter Plot')
plt.xlabel('day')
plt.ylabel('tip')

plt.scatter(data_tips['day'], data_tips['tip'])
plt.show()
```

### Latihan Pada Modul

```python
# Menghitung persentase
plot_data = data_tips['sex'].value_counts(normalize=True).reset_index()
plot_data.columns = ['Gender', 'Percentage']
plot_data['Percentage'] *= 100

# Membuat visualisasi
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Gender', y='Percentage', data=plot_data, palette=['skyblue', 'pink'])

plt.title('Persentase Pemberi TIP')
plt.ylabel('Persentase (%)')
plt.ylim(0, 100)
plt.show()
```

### Visualisasi Data Student Performance

```python
data = pd.read_csv('StudentsPerformance.csv')
data.head()
```

```python
data.info()
```

```python
# Hitung rata-rata nilai math berdasarkan gender
nilai_rata_rata = data.groupby("gender")["math score"].mean()

# Membuat bar chart
plt.bar(nilai_rata_rata.index, nilai_rata_rata.values)

# Judul dan label
plt.title("Rata-rata Nilai Math Berdasarkan Jenis Kelamin")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Rata-Rata Nilai Matematika")

plt.show()
```
